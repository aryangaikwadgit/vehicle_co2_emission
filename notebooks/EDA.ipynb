```python
categorical_feature_df =  df.select_dtypes(include = "object").drop(columns = ["Make","Model"])
categorical_feature = categorical_feature_df.columns.tolist()
categorical_feature
```

```python
numerical_features_df =  df.select_dtypes(include = "number").drop(columns = ["Co2_Emissions(G/Km)"])
numerical_features =numerical_features_df.columns.tolist() 
numerical_features
```

```python
# skewness
df[numerical_features].describe()
```

```python
def statistical_analysis(x):
    stats={
        "mean" : x.mean(),
        "median" : x.median(),
        "mode" : x.mode().iloc[0],
        "var" : x.var(),
        "std" :x.std(),
        "skewness" : x.skew()
    }
    return pd.DataFrame(stats)
statistical_analysis(df[numerical_features]).sort_values(by = "mean")
```

```python
plt.figure(figsize=(12,12))
r,c,q=3,2,1
for x in numerical_features:
    plt.subplot(r,c,q)
    sns.histplot(df[x],bins = 30,kde = True)
    plt.title(f"Histogram of {x}")
    q = q+1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,15))
r,c,q=3,2,1
for x in numerical_features:
    plt.subplot(r,c,q)
    sns.boxplot(x=df[x])
    plt.title(f"Histogram of {x}")
    q = q+1

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for x in categorical_feature:
    sns.barplot(y=df[x].values,x=df[x].index)
    plt.title(f"Histogram of {x}")
    plt.show()
plt.tight_layout()
```
